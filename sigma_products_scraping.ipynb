```python
import requests
res = requests.get("https://www.sigma-computer.com/category?id=2")
res
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, "html.parser")
len(soup.find_all("div", attrs={"class":"product-layout col-lg-15 col-md-4 col-sm-6 col-xs-6"}))
```

```python
products_tag = soup.find_all("div", attrs={"class":"product-layout col-lg-15 col-md-4 col-sm-6 col-xs-6"})
```

```python
#product_link, product_description, price, availability, brand, model, type, viewes
#product_link
url = "https://www.sigma-computer.com/"
product_link = url + products_tag[8].find("div", attrs={"class":"right-block"}).find("a").attrs['href']
product_link
```

```python
#product_description
product_description = products_tag[8].find("div", attrs={"class":"right-block"}).find("a").text
product_description
```

```python
#price 
product_price = int(products_tag[8].find("div", attrs={"class":"right-block"}).find("span", attrs={"class":"price-new"}).text)
product_price
```

```python
#data from inside one produt
res1 = requests.get(product_link)
soup1 = BeautifulSoup(res1.text, "html.parser")
```

```python
#availability 
if soup1.find("div", attrs={"class":"cart"}).find("button", attrs={"data-toggle":"tooltip"}).text.strip("\n ").lower() == "add to cart" :
    product_available = "In Stock"
else:
    product_available = "out of stock"
product_available
```

```python
#brand
product_brand = soup1.find("div", attrs={"class":"brand"}).find("a").text.strip()
product_brand
```

```python
#model
product_model = soup1.find("div", attrs={"class":"model"}).find("a").text
product_model
```

```python
#type
product_type = soup1.find("ul", attrs={"class":"breadcrumb"}).find_all("li")[2].text
product_type
```

```python
#viewes
product_viewes = int(soup1.find("div", attrs={"class":"inner-box-desc"}).find_all("div")[3].find("h3").text.split()[2])
product_viewes
```

```python
#scrap number of pages
num_of_pages = int(soup.find("ul", attrs={"class":"pagination"}).find_all("li")[-2].text)
num_of_pages
```

```python
with open("sigma_products_scraping.csv", mode="w", encoding="utf-8") as fd:
    #header
    fd.write("Product_Type,Brand,Model,Product_Description,Price,Viewes,Availability,Product_Link\n")
    
    #scrap number of pages
    res_n_p = requests.get("https://www.sigma-computer.com/category?id=2")
    soup_n_p = BeautifulSoup(res_n_p.text, "html.parser")
    num_of_pages = int(soup_n_p.find("ul", attrs={"class":"pagination"}).find_all("li")[-2].text)

    #loop_pages
    for page in range(num_of_pages):
        try:
            print(f'Scraping page {page + 1}')
            res2 = requests.get(f"https://www.sigma-computer.com/category?id=2&page={page + 1}")
            soup2 = BeautifulSoup(res2.text, "html.parser")
            products_tag = soup.find_all("div", attrs={"class":"product-layout col-lg-15 col-md-4 col-sm-6 col-xs-6"})
            
            #content
            for i in range(len(products_tag)):
                print(f'Scraping product {i + 1}')
                #product_link
                url = "https://www.sigma-computer.com/"
                product_link = url + products_tag[i].find("div", attrs={"class":"right-block"}).find("a").attrs['href']

                #product_description
                product_description = products_tag[i].find("div", attrs={"class":"right-block"}).find("a").text.replace('"',"")

                #price 
                product_price = int(products_tag[i].find("div", attrs={"class":"right-block"}).find("span", attrs={"class":"price-new"}).text)

                #data from inside one produt
                res1 = requests.get(product_link)
                soup1 = BeautifulSoup(res1.text, "html.parser")

                #availability
                if soup1.find("div", attrs={"class":"cart"}).find("button", attrs={"data-toggle":"tooltip"}).text.strip("\n ").lower() == "add to cart" :
                    product_available = "In Stock"
                else:
                    product_available = "out of stock"  
                                
                #brand
                product_brand = soup1.find("div", attrs={"class":"brand"}).find("a").text.strip()

                #model
                product_model = soup1.find("div", attrs={"class":"model"}).find("a").text

                #type
                product_type = soup1.find("ul", attrs={"class":"breadcrumb"}).find_all("li")[2].text

                #viewes
                product_viewes = int(soup1.find("div", attrs={"class":"inner-box-desc"}).find_all("div")[3].find("h3").text.split()[2])

                fd.write(f'\"{product_type}\",\"{product_brand}\",\"{product_model}\",\"{product_description}\",\"{product_price}\",\"{product_viewes}\",\"{product_available}\",\"{product_link}\"\n')

        except:
            continue

```
