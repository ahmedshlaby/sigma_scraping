```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://www.sigma-computer.com/search?search=laptop&submit_search=&route=product%2Fsearch")
soup = BeautifulSoup(res.text, "html.parser")
```

```python
len(soup.find_all("div", attrs={"class":"product-layout col-lg-15 col-md-4 col-sm-6 col-xs-12"}))
```

```python
#scrap all products tags
products_tags = soup.find_all("div", attrs={"class":"product-layout col-lg-15 col-md-4 col-sm-6 col-xs-12"})
```

```python
#product_link, product_description, price, availability, brand, model, type, viewes
#product_link
url = "https://www.sigma-computer.com/"
product_link = url + products_tags[43].find("div", attrs={"class":"product-image-container second_img"}).find("a").attrs['href']
product_link
```

```python
#product_description
product_description = products_tags[43].find("div", attrs={"class":"right-block"}).find("div", attrs={"class":"caption hide-cont"}).find("h4").text.replace(",","")
product_description
```

```python
#availability
product_available = products_tags[0].find("div", attrs={"class":"row"}).find_all("div", attrs={"class":"col-md-6"})[0].text.strip("\n ●")
product_available
```

```python
#price : old_price,new_price
new_price = int(products_tags[0].find("div", attrs={"class":"row"}).find_all("div", attrs={"class":"col-md-6"})[1].find_all("span")[0].text.split()[0])
old_price = int(products_tags[0].find("div", attrs={"class":"row"}).find_all("div", attrs={"class":"col-md-6"})[1].find_all("span")[1].text.split()[0])
old_price
```

```python
#brand
res_p = requests.get(product_link)
soup_p = BeautifulSoup(res_p.text, "html.parser")
product_brand = soup_p.find("div", attrs={"class":"inner-box-desc"}).find("div", attrs={"class":"brand"}).find("a").text.strip()
product_brand
```

```python
#model
product_model = soup_p.find("div", attrs={"class":"inner-box-desc"}).find_all("div", attrs={"class":"model"})[0].find("a").text.strip()
product_model
```

```python
#viewes
product_viewed = int(soup_p.find("div", attrs={"class":"inner-box-desc"}).find_all("div", attrs={"class":"model"})[2].find("h3").text.split(":   ")[1].split()[0])
product_viewed
```

```python
#Product_Type
if product_brand.lower() in ["dell", "lenovo", "acer", "hp", "msi", "asus", "apple", "samsung", "lg", "alienware", "microsoft", "gigabyte"]:
    produt_type = "Laptop"
else:
    produt_type = "Other"
produt_type
```

```python
with open("products_sigma.csv", mode="w", encoding="utf-8") as fd:
    #header
    fd.write("Product_Type,Brand,Model,Product_Description,new_price,old_price,Viewes,Availability,Product_Link\n")

    #content
    for i in range(len(products_tags)):
        try:
            print(f'scarping product {i + 1}')
            #product_link
            url = "https://www.sigma-computer.com/"
            product_link = url + products_tags[i].find("div", attrs={"class":"product-image-container second_img"}).find("a").attrs['href']

            #product_description
            product_description = products_tags[i].find("div", attrs={"class":"right-block"}).find("div", attrs={"class":"caption hide-cont"}).find("h4").text.replace(",","")
        
            #availability
            product_available = products_tags[i].find("div", attrs={"class":"row"}).find_all("div", attrs={"class":"col-md-6"})[0].text.strip("\n ●")
            
            #price : old_price,new_price
            new_price = int(products_tags[i].find("div", attrs={"class":"row"}).find_all("div", attrs={"class":"col-md-6"})[1].find_all("span")[0].text.split()[0])
            old_price = int(products_tags[i].find("div", attrs={"class":"row"}).find_all("div", attrs={"class":"col-md-6"})[1].find_all("span")[1].text.split()[0])

            #brand
            res_p = requests.get(product_link)
            soup_p = BeautifulSoup(res_p.text, "html.parser")
            product_brand = soup_p.find("div", attrs={"class":"inner-box-desc"}).find("div", attrs={"class":"brand"}).find("a").text.strip()

            #model
            product_model = soup_p.find("div", attrs={"class":"inner-box-desc"}).find_all("div", attrs={"class":"model"})[0].find("a").text.strip()
            
            #viewes
            product_viewed = int(soup_p.find("div", attrs={"class":"inner-box-desc"}).find_all("div", attrs={"class":"model"})[2].find("h3").text.split(":   ")[1].split()[0])

            #Product_Type
            if product_brand.lower() in ["dell", "lenovo", "acer", "hp", "msi", "asus", "apple", "samsung", "lg", "alienware", "microsoft"]:
                produt_type = "Laptop"
            

            fd.write(f'\"{produt_type}\",\"{product_brand}\",\"{product_model}\",\"{product_description}\",\"{new_price}\",\"{old_price}\",\"{product_viewed}\",\"{product_available}\",\"{product_link}\"\n')
        except:
            continue
```
